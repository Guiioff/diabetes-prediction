```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = "diabetes_dataset.csv"

df = pd.read_csv(path)
```

```python
numeric_cols = ['year', 'age', 'bmi', 'hbA1c_level', 'blood_glucose_level']
cat_cols = ['gender', 'smoking_history', 'diabetes']
```

```python
plt.figure(figsize=(9,5))

for i, col in enumerate(numeric_cols, 1):
  plt.subplot(2,3,i)
  sns.kdeplot(df[col], fill=True)
  plt.title(f'{col}')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(9,5))

for i, col in enumerate(cat_cols, 1):
  plt.subplot(2,3,i)
  sns.histplot(df[col], fill=True)
  plt.title(f'{col}')
  plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
